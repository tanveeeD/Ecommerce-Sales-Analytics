```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset

df = pd.read_csv("../excel/SuperstoreSales.csv")
```

```python
# Query 1 : Dataset Shape

# Business Question: How many rows and columns are present?

df.shape
```

```python
# Query 2 : Display First 5 Records

# Business Question: Preview the dataset.

df.head()
```

```python
# Query 3 : Dataset Information

# Business Question: What are the data types and non-null values?

df.info()
```

```python
# Query 4 : Statistical Summary

# Business Question: What are the statistical details of numerical columns?

df.describe()
```

```python
# Query 5 : Missing Values

# Business Question: Are there any missing values in the dataset?

df.isnull().sum()
```

```python
# Query 6 : Duplicate Records

# Business Question: Are there any duplicate records?

df.duplicated().sum()
```

```python
# Query 7 : Total Sales

# Business Question: What is the total sales?

df["Sales"].sum()
```

```python
# Query 8 : Total Profit

# Business Question: What is the total profit?

df["Profit"].sum()
```

```python
# Query 9 : Total Orders

# Business Question: How many unique orders are present?

df["Order ID"].nunique()
```

```python
# Query 10 : Total Customers

# Business Question: How many unique customers are present?

df["Customer ID"].nunique()
```

```python
# Query 11 : Sales by Category

# Business Question: Which category generates the highest sales?

sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

sales

plt.figure(figsize=(6,4))
sales.plot(kind="bar", color="skyblue")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
# Query 12 : Profit by Category

# Business Question: Which category generates the highest profit?

profit = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)

profit
plt.figure(figsize=(6,4))
profit.plot(kind="bar", color="skyblue")
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.xticks(rotation=0)
plt.show()
```

```python
# Query 13 : Sales by Sub-Category

# Business Question: Which sub-category generates the highest sales?

subcat_sales = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)

subcat_sales
plt.figure(figsize=(10,5))
subcat_sales.plot(kind="bar", color="teal")
plt.title("Sales by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Query 14 : Profit by Sub-Category

# Business Question: Which sub-category generates the highest profit?

subcat_profit = df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False)

subcat_profit
plt.figure(figsize=(10,5))
subcat_profit.plot(kind="bar", color="lightgreen")
plt.title("Profit by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.show()
```

```python
# Query 15 : Sales by Region

# Business Question: Which region contributes the highest sales?

region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

region_sales
plt.figure(figsize=(6,4))
region_sales.plot(kind="bar", color="orange")
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
# Query 16 : Profit by Region

# Business Question: Which region generates the highest profit?

region_profit = df.groupby("Region")["Profit"].sum().sort_values(ascending=False)

region_profit
plt.figure(figsize=(6,4))
region_profit.plot(kind="bar", color="lightpink")
plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.xticks(rotation=0)
plt.show()
```

```python
# Query 17 : Top 10 States by Sales

# Business Question: Which states have the highest sales?

state_sales = df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)

state_sales
plt.figure(figsize=(10,5))
state_sales.plot(kind="bar", color="royalblue")
plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Query 18 : Top 10 Customers by Sales

# Business Question: Which customers generated the highest sales?

top_customers = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)

top_customers
plt.figure(figsize=(10,5))
top_customers.plot(kind="bar", color="crimson")
plt.title("Top 10 Customers by Sales")
plt.xlabel("Customer Name")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Query 19 : Top 10 Products by Sales

# Business Question: Which products generated the highest sales?

top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

top_products
plt.figure(figsize=(12,5))
top_products.plot(kind="bar", color="darkcyan")
plt.title("Top 10 Products by Sales")
plt.xlabel("Product Name")
plt.ylabel("Sales")
plt.xticks(rotation=90)
plt.show()
```

```python
# Query 20 : Monthly Sales Trend

# Business Question: What are the monthly sales trends?

df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)

df["Month"] = df["Order Date"].dt.month_name()

month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

monthly_sales = (
    df.groupby("Month")["Sales"]
      .sum()
      .reindex(month_order)
)

monthly_sales
```

```python
# Query 21 : Correlation Heatmap

# Business Question: How are numerical columns related?

plt.figure(figsize=(8,6))

sns.heatmap(
    df.select_dtypes(include=np.number).corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
# Query 22 : Profit Margin Analysis

# Business Question: What is the profit margin for each order?

df["Profit Margin"] = (df["Profit"] / df["Sales"]) * 100

df[["Sales", "Profit", "Profit Margin"]].head()
```

```python
# Query 23 : Average Profit Margin by Category

# Business Question: Which category has the highest average profit margin?

profit_margin = df.groupby("Category")["Profit Margin"].mean().sort_values(ascending=False)

profit_margin
plt.figure(figsize=(6,4))
profit_margin.plot(kind="bar", color="mediumseagreen")
plt.title("Average Profit Margin by Category")
plt.xlabel("Category")
plt.ylabel("Profit Margin (%)")
plt.xticks(rotation=0)
plt.show()
```

```python
# Query 24 : Discount vs Profit

# Business Question: Does a higher discount reduce profit?

plt.figure(figsize=(8,5))
plt.scatter(df["Discount"], df["Profit"], color="crimson")
plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
```

```python
# Query 25 : Monthly Profit Trend

# Business Question: How does profit vary across different months?

df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Month"] = df["Order Date"].dt.month_name()

month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

monthly_profit = (
    df.groupby("Month")["Profit"]
      .sum()
      .reindex(month_order)
)

monthly_profit
```

```python
plt.figure(figsize=(10,5))
monthly_profit.plot(kind="line", marker="o", color="darkgreen")
plt.title("Monthly Profit Trend")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```

```python
# Query 26 : Shipping Mode Analysis

# Business Question: Which shipping mode is used most frequently?

ship_mode = df["Ship Mode"].value_counts()

ship_mode
plt.figure(figsize=(6,4))
ship_mode.plot(kind="bar", color="cornflowerblue")
plt.title("Orders by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Number of Orders")
plt.xticks(rotation=0)
plt.show()
```

```python
# Query 27 : Pairplot

# Business Question: What is the relationship between Sales, Profit and Discount?

sns.pairplot(
    df[["Sales", "Profit", "Discount"]]
)

plt.show()
```

```python
# Query 28 : Correlation Matrix

# Business Question: How are numerical variables correlated?

correlation = df[["Sales", "Profit", "Discount", "Quantity"]].corr()

correlation
plt.figure(figsize=(6,5))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

# Final Business Insights

- Technology generated the highest sales.
- Technology generated the highest profit.
- Furniture generated high sales but comparatively lower profit.
- The West region generated the highest sales.
- Standard Class was the most frequently used shipping mode.
- Higher discounts often reduced profit.
- A small group of customers contributed a significant share of total revenue.
